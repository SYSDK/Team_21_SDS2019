```python
import pandas as pd
import requests
import time

for i in range(1, 59):
    if i == 1:
        response = requests.get('https://www.the-numbers.com/movie/budgets/all')
    else:
        response = requests.get('https://www.the-numbers.com/movie/budgets/all/' + str(i) + '01')
    html = response.text
    
    tables = pd.read_html(html)
    table = tables[0]
    table.columns = ['toDrop', 'releaseDate', 'title', 'budget', 'boDomestic', 'boWorldwide']
    table = table.drop('toDrop', axis = 1)
    
    if i == 1:
        df = table.copy()
    else:
        df = df.append(table)
        
    print(i)
    time.sleep(1)

df = df.reset_index()
df.to_pickle('budgets_raw.pkl')
```

```python
def last_four(string):
    try:
        return int(string[-4:])
    except:
        return None
    
def numerize_money(string):
    if isinstance(string, str):
        try:
            return int(float(string.replace('$', '').replace(',', '')))
        except:
            return None
    else:
        return None
    
def fix_title(string):
    return string.replace('Ã¢\x80\x99', '\'')

df['year'] = df['releaseDate'].apply(last_four)
df['budget'] = df['budget'].apply(numerize_money)
df['title'] = df['title'].apply(fix_title)
df = df.drop(['index', 'boDomestic', 'boWorldwide'], axis = 1)
```

```python
df.to_pickle('budgets_clean.pkl')
```
